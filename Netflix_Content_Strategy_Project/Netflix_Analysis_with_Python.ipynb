# Netflix Content Strategy Analysis

This notebook analyzes Netflix's content strategy based on viewership data for the year 2023. We explore viewership trends by content type, language, release month, and more to derive insights into strategic decisions.


```python
# Importing all necessarylibraries for this project
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the netflix dataset
netflix_data = pd.read_csv("netflix_content_2023.csv")
netflix_data.head()
```

## Data preparation
* Convert Hours viewed to numeric


```python
netflix_data['Hours Viewed (Converted)'] = netflix_data['Hours Viewed'].replace(',', '', regex=True).astype(float)
netflix_data[['Title', 'Hours Viewed (Converted)']].head()
```

```python
netflix_data['Hours Viewed (Converted 2)'] = netflix_data['Hours Viewed'].astype(str).str.replace(r'[^0-9.\-]', '', regex=True).astype('float64') #astype(str) ensure string; .str.replace(r'[^0-9.\-]', '', regex=True) keep digits, dot, minus
netflix_data[['Title', 'Hours Viewed (Converted 2)']].head()
```
